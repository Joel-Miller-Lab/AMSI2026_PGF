# MultiType Galton-Watson Processes

We now consider what happens when we have multi-type Galton-Watson processes.  Here an individual of one type can give birth to individuals of multiple types.  

The number of offspring an individual will have may depend on its own type, but it is independent of the outcomes of any other individuals.

We will be interested in determining many of the same quantities we focused on for a single-type Galton-Watson process.  In addition, we will want to be able to calculate quantities related to the joint distributions.
