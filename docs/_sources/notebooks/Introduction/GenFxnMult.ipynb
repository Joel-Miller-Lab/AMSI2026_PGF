# Multiplication of Generating Functions

We'll see that there are several uses for multiplication of two generating functions.  First, we'll look at the sequence of numbers that results from multiplying two generating functions together and then we'll look at some applications.

Let's assume we have two generating functions: 

\begin{align*}
f(x) &= f_0 + f_1 x + f_2 x^2 + \cdots\\
g(x) &= g_0 + g_1 x + g_2 x^2 + \cdots
\end{align*}

We can represent multiplication of the sums through a table:

|||$f_0$|$f_1x$|f_2x^2$|$f_3x^3$|$f_4x^4$|$f_5x^5$|$\cdots$|
|--|--|--|--|--|--|--|--|--|
|||||||||
|$g_0$||$f_0g_0$|$f_1g_0x$|$f_2g_0x^2$|$f_3g_0x^3$|$f_4g_0x^4$|$f_5g_0x^5$|$\cdots$|
|$g_1x$||$f_0g_1x$|$f_1g_1x^2$|$f_2g_1x^3$|$f_3g_1x^4$|$f_4g_1x^5$|$f_5g_1x^6$|$\cdots$|
|$g_2x^2$||$f_0g_2x^2$|$f_1g_2x^3$|$f_2g_2x^4$|$f_3g_2x^5$|$f_4g_2x^6$|$f_5g_2x^7$|$\cdots$|
|$g_3x^3$||$f_0g_3x^3$|$f_1g_3x^4$|$f_2g_3x^5$|$f_3g_3x^6$|$f_4g_3x^7$|$f_5g_3x^8$|$\cdots$|
|$g_4x^4$||$f_0g_4x^4$|$f_1g_4x^5$|$f_2g_4x^6$|$f_3g_4x^7$|$f_4g_4x^8$|$f_5g_4x^9$|$\cdots$|


By looking at this table, we can see that the coefficient of $x^n$ is equal to the sum of all products of the form $f_k g_{n-k}$.  That is,

$$
f(x) g(x) = \sum_{n=0}^\infty \left(x^n\sum_{k=0}^n f_kg_{n-k}\right) 
$$

## Special case: $f(x) = 1 + x + x^2 + \cdots$

An important special case occurs when $f(x) = 1 + x + x^2 + \cdots$.  Then

$$
f(x)g(x) = g_0 + (g_0+g_1) x + (g_0+g_1+g_2)x^2 + \cdots
$$
This is a generating function whose coefficients are $g_0$, $g_0+g_1$, $g_0+g_1+g_2$, $\ldots$.  These are the partial sums of the sequence $g_0, g_1, \ldots$.
So given a generating function $g(x)$, we can construct the generating function of the partial sums by multiplying by $f(x)$.
