# Forward Kolmogorov Equations

These notes introduce the **forward Kolmogorov equations** for continuous-time Markov processes.  They play the same structural role for continuous time that recursion for $\Phi_g(x)$ played for discrete-time processes in earlier sections.  


## Continuous-time Markov processes


Let $(X_t)_{t\ge 0}$ be a stochastic process taking values in a countable state space.

The process is a **continuous-time Markov chain (CTMC)** if

$$
\mathbb{P}(X_{t+s}=j \mid X_t=i, \text{past}) = \mathbb{P}(X_{t+s}=j \mid X_t=i).
$$

Define

$$
p_{ij}(t) = \mathbb{P}(X_t=j \mid X_0=i).
$$

Let $P(t)=[p_{ij}(t)]$. Then:
- $P(0)=I$
- $P(t+s)=P(t)P(s)$ (Chapman�Kolmogorov)

## 2. The generator

For $i\neq j$, let $q_{ij}$ be the rate of jumps from $i$ to $j$.
Define

$$
q_{ii}=-\sum_{j\neq i}q_{ij}.
$$

For small $h$:


$$
p_{ij}(h)=q_{ij}h+o(h), \quad i\neq j,
$$

$$
p_{ii}(h)=1+q_{ii}h+o(h).
$$

## 3. Forward Kolmogorov equations

From $P(t+h)=P(t)P(h)$,
\[
\frac{P(t+h)-P(t)}{h}=P(t)\frac{P(h)-I}{h}.
\]

Letting $h\downarrow0$ gives
\[
\boxed{\frac{d}{dt}P(t)=P(t)Q.}
\]

Component form:
\[
\boxed{\frac{d}{dt}p_{ij}(t)=\sum_k p_{ik}(t)q_{kj}.}
\]

## 4. Interpretation

Each term $p_{ik}(t)q_{kj}$ represents **probability flow** from state $k$ into state $j$.

This has the same gain�loss structure as discrete-time recursions and Galton�Watson generation updates.

## 5. Example: pure birth process

Let $q_{n,n+1}=\lambda$ and $q_{n,n}=-\lambda$.

Then
\[
\frac{d}{dt}p_{in}(t)=\lambda p_{i,n-1}(t)-\lambda p_{in}(t).
\]

## 6. Connection to branching processes

Forward Kolmogorov equations are the **continuous-time analogue** of the recursion relations
used earlier for PGFs and Galton�Watson processes.
