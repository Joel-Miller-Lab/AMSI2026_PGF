# The Backward Kolmogorov Equation

The backward Kolmogorov equation along similar lines to the forward Kolmogorov equation.  We assume we know $\Phi(x,t)$ and $\hat{\mu}(x)$ and we want to determine $\Phi(x, t + \Delta t)$.

However, instead of taking the PGF for time $t$, $\Phi(x,t)$ and stepping forward an amount $\Delta t$ using $\hat{\mu}(x)$ as we did in deriving the Forward Kolmogorov Equations, we take the PGF for time $\Delta t$ from {prf:ref}`lem-PGFatDt` and step forward an amount $t$ using $\Phi(x,t)$.  This will arrive at a different partial differential equation for $\Phi(x,t)$.  However, both equations will yield the same solution $\Phi(x,t)$.  Each equation gives insight into different properties of $\Phi(x,t)$.



Given $\Phi(x,0)= x$, that is, there is one individual at time $t=0$, {prf:ref}`lem-PGFatDt` tells us

$$
\Phi(x,\Delta t) = (1-r\Delta t)x + r\Delta t \hat{\mu}(x)
$$

To determine $\Phi(x,\Delta t + t)$, we now use {prf:ref}`lem-PhiTimeSum` with $T_1=\Delta t$ and $T_2=t$.  So

$$
\Phi(x,t+\Delta t) = (1-r\Delta t) \Phi(x,t) + r \Delta t \hat{\mu}(\Phi(x,t))
$$
