# Generating Functions

## Introduction
In this unit, we will study how probability generating functions (PGFs) can be used in biological contexts.  This requires learning 
- the mathematical theory of PGFs, 
- how to convert the biological applications into the mathematical language of PGFs, and 
- how to interpret the mathematical results in terms of the biological applications.

This has created a conundrum for me as I write these notes.  Pedagogically, I believe it is best to teach the material so that the application and the relevant mathematical theory are taught at the same time.  But should I write these notes in that order, or should I keep the mathematical theory and the applications separate?  I have finally decided to separate the material so that when we get into later applications it is easier to find the mathematical results in one place.  

So this chapter is structured as a self-contained reference to probability generating functions, focusing on those results which we need in our applications.  The main material of the subject begins in {numref}`Chapter %s <sec:GaltWatsonIntro>`, with frequent references back to this material, and guidance about what sections of this chapter to read simultaneously.
