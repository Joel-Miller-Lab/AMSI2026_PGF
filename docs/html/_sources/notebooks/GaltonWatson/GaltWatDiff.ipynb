# Differentiation and the Galton Watson Process

Please read {numref}`sec:GenFxnsDiff` prior to reading this section.

Consider A Galton-Watson process.  Let $\mu(x)$ be the PGF of the offspring distribution. We have learned that the extinction probability $\alpha$ is a solution to the equation $x = \mu(x)$.

In the plots given in  {numref}`sec:GaltWatExtinct`, we saw two solutions to $x=\mu(x)$.  One was between $0$ and $1$ and another was equal to $1$.  We will learn later that if there is a solution $x_0$ with $x_0 = \mu(x_0)$ and $0\leq x_0 <1$, that will be the correct solution for $\alpha$.  Otherwise, $\alpha=1$ will be the correct solution.

For now, we'll try to find conditions for which there is guaranteed to be a solution $0 \leq \alpha <1$  and for which there no such solution.  This will determine whether $\alpha=1$ in which case extinction is guaranteed or $\alpha<1$ in which case long-term persistence is possible.

Intuitively, we might expect that extinction is guaranteed if the expected number of offspring is less than $1$.  Our results about the moments of $\mu(x)$ in {numref}`sec:GenFxnsDiff` show that the expected number of offspring is $\mu'(1)$.  This observation provides us a mathematical handle that we can use to produce a rigorous proof verifying our expection.

Before attempting the proof, we plot the line $y=x$ and the curves $y = \mu_\lambda(x)$ for a Poisson distribution with mean $\lambda$ to strengthen our intuition.

````python
import matplotlib.pyplot as plt 
import numpy as np

def mu(x,lamb):  #The PGF for a Poisson distribution of mean lambda
    return np.exp(lamb*(x-1))

x = np.linspace(0,1,100)
for lamb in [0.5,1,2,3]:
    plt.plot(x, mu(x,lamb), label='$\\mu_{' + str(lamb) + '}(x)$')
plt.plot(x, x, color = 'k', label = '$y=x$')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.legend()

plt.show()
````


We see that for these cases each $\mu(x)$ is a convex function, and $\mu(0)>0$.  As noted before $\mu(1)=1$.  The plots that have slope greater than $1$ at $x=1$ have $\mu(x)<x$ for sufficiently large values of $x<1$, and then they cross $y=x$ at some smaller value of $x$.  The plots that have slope less than (or equal to) $1$ at $x=1$ have a shallower slope for $x<1$, and so they remain above $y=x$ for smaller values of $x$.

Now we'll try to make these observations more general beyond the handful of plots shown here.  To avoid the awkward case of $\mu(x)=x$ we will focus just on $\mu'(1)<1$ or $\mu'(1)>1$.

We first look at the case where $\mu'(1) < 1$, then $\mu(x)-x$ is positive for values of $x$ close enough to but less than $1$.  Since $\mu(x)$ is a polynomial with non-negative coefficients, $\mu'(x) \leq \mu'(1)$ for $0< x < 1$, so $\mu(x)-x$ is a decreasing function in this interval.  As it equals $0$ at $x=1$, it must be positive for $0 \leq x <1$.  So there is no other solution in $0 \leq x <1$.

However if $\mu'(1) > 1$, then $\mu(x) - x<0$ for $x<1$ close enough to $1$.  Further $\mu(0)$ must be non-negative so $\mu(x)-x \geq 0$ at $x=0$.  In the case that $\mu(x)-x=0$ at $x=0$, we have found a solution $x = \mu(x)$ at $x=0$.  In the case that $\mu(x)-x>0$ at $x=0$, we note that $\mu(x)-x$ is a continuous function that is positive at $x=0$ and negative for large enough $x<1$.  Thus there must be some point $\alpha$ with $0<\alpha<1$ where $\mu(\alpha)-\alpha=0$, that is $\alpha=\mu(\alpha)$.

Thus we see that if $\mu'(1)<1$ then extinction is guaranteed ($\alpha=1$ is the only solution), while if $\mu'(1)>1$, there is a solution $0\leq \alpha<1$ which means that extinction is not guaranteed.

As we have noted above, $\mu'(1)$ is the average of the distribution, that is the expected number of offspring.  So extinction is guaranteed if the average number of offspring is less than $1$, but if the expected number is greater than $1$, then long-term persistence is possible.

We have proven a Theorem:
```{prf:theorem} Galton-Watson extinction threshold
:label: thm-GWExtinctThresh
Given a Galton-Watson process whose offspring distribution has PGF $\mu(x)$:
- If $\mu'(1)<1$, then extinction is guaranteed.  
- If $\mu'(1)>1$, then long-term persistence (that is, never going extinct) is possible.
```
In the case of long-term persistence, we typically see exponential growth.

As presented, the theorem did not address $\mu'(1)=1$.  It is possible to handle this possibility.  We need to separate it into two cases: $\mu(x)=x$ and $\mu(x) \neq x$. If $\mu(x)=x$, then long-term persistence is guaranteed.  If $\mu'(1)=1$ and $\mu(x) \neq x$, then extinction is guaranteed (but it may take a very long time).
