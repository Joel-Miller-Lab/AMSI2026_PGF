# A paradigm change

Most functions that we encounter in our mathematical training are smooth and infinitely differentiable.  In this case, they have a Taylor Series:

$$
f(x) = c_0 + c_1 x + c_2 x^2 + \cdots
$$
where $c_0 = f(0)$, $c_1 = f'(0)$, $c_2 = f''(0)/2$, $\ldots$, $c_n = f^{(n)}(0)/n!, \ldots$.  

Usually, when we think about a function $f(x)$, we are interested in the values it produces for a given input $x$.  If asked to visualize a function $y = (1 + x^2)/2$, many of us will think of the parabola plotted below.



````python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-2,2,100)
plt.plot(x, (1+x**2)/2)
plt.xlabel('$x$')
plt.ylabel('$y$')
````

Very few of us would instead think of the sequence of numbers: $c_0=1/2$, $c_1=0$, $c_2=1/2$, $c_3=0$, $c_4=0$, $\ldots$.

```{prf:remark} Function and Taylor Series: the usual approach
:label: remark-fxn2TS

Usually, the values of the function form the ``concrete object'' we want to know about.  The Taylor Series is mostly useful to help us calculate the values of the function.
```


We are going to switch our focus.  For our purposes, we will be primarily interested in the coefficients of the Taylor Series.  We will only care about the values of the function when we can use them to learn about properties of the sequence of coefficients.

```{prf:remark} Function and Taylor Series: the generating function approach
:label: remark-TS2fxn

In the paradigm we will use, the sequence of coefficients of the Taylor Series is the ``concrete object'' we want to study.  The values of the function are mostly useful for what they tell us about the sequence.
```



## Definitions
Our focus on the sequence of coefficients of a function's Taylor Series expansion motivates the following definition:

```{prf:definition} Generating Function
:label: defn-GenFxn

Given a seqence of numbers $c_0$, $c_1$, $\ldots$, the corresponding generating function is the function

$$
f(x) = \sum_{n=0}^\infty c_n x^n
$$
is the **Generating Function** of the sequence.
```
Sometimes we will call this an **Ordinary Generating Function** or **OGF**.

Our goal in creating a generating function is to apply tools we learn in other areas of mathematics to the function in order to learn properties of the sequence and the process that creates that sequence.



We will discover later that when the coefficients come from a probability distribution of non-negative integers the operations of adding, multiplying, and composing generating functions correspond to biological properties related to reproduction.  This motivates the study of a special kind of generating function:

```{prf:definition} Probability Generating Function
:label: defn-PGF

Given a probability distribution for non-negative integers: $p_0, p_1, \ldots,$ with $p_n \geq 0$ for all $n$ and $\sum p_n = 1$, the function

$$
\mu(x) = \sum_{n=0}^\infty p_n x^n
$$
is the **Probability Generating Function (PGF)** of the distribution.
```

So a PGF is simply an OGF whose coefficients correspond to a probability distribution for non-negative integers.

## Two famous quotes

To illustrate some of the ideas underlying how we use generating functions, here are two famous quotes about them, illustrated using the Taylor Series $e^x = 1 + x + \frac{1}{2} x^2 + \frac{1}{6}x^3 + \cdots$:


*"A generating function is a clothesline on which we hang up a sequence of numbers for display."*  H. Wilf

This "clothesline" is an organized way of displaying a sequence of numbers.  However, one of the uses of generating functions is that we can often rewrite the generating function in a compact way, which is captured in the following:

*"A generating function is a device somewhat similar to a bag. Instead of carrying many little objects detachedly, which could be embarrassing, we put them all in a bag, and then we have only one object to carry, the bag."*   George Pólya
