# Infectious Diseases 

## Introduction
We will consider directly-transmitted infectious diseases.  Interactions between infectious and susceptible individuals can lead to infection of the susceptible individual.

There are two general classes of infectious disease that are typically studied: 
- **SIS Diseases** for which susceptible individuals can become infected, and later recover and are susceptible to infection again.
- **SIR Diseases** which transmit similarly to SIS diseases, but once an individual recovers he/she gains permanent immunity.


Typically we will use either a continuous-time approach in which events occur with rates or a discrete-time approach in which events are separated by generations.

### Continuous-time infectious disease models
When we consider the continuous-time disease model, infected individuals transmit with rate $\beta$ and recover with rate $\gamma$.  
- When they transmit, a single individual is chosen uniformly at random from the population to receive the transmission.  
  - If the recipient is susceptible, which occurs with probability $S(t_i)/N$, then an infection occurs.  For both SIS and SIR:
    - $I(t)$ jumps so that $I(t_{i+1}) = I(t_i)+1$ and 
    - $S(t)$ jumps to $S(t_{i+1}) = S(t_i)-1$.  
    
    If the recipient is not susceptible nothing happens.
- When they recover, they cease to be infectious.  
  - In the SIS case, they are immediately susceptible again:
    - $I(t_{i+1})= I(t_i)-1$ and 
    - $S(t_{i+1})=S(t_i)+1$, 

    while 
  - in the SIR case they are immune to future infection 
    - $I(t_{i+1})= I(t_i)-1$ and 
    - $R(t_{i+1}) = R(t_i)+1$.

In between the jump times $t_0$, $t_1$, $\ldots$, all of the counts remain constant.




### Generation-based infectious disease models

In the generation-based approach, we assume the total number of transmissions caused by an infectious individual during its infectious period is a random number, and its distribution is the *offspring distribution*.  We will typically use $\mu(x)$ as the PGF of the offspring distribution.  

Note that if some recipients of transmissions are not susceptible, not all transmissions will cause infections.  Each transmission is successful with probability $S/N$.

We will assume the population is *well-mixed*, that is, the recipient of any transmission is randomly selected from the population.



### Conceptual reduction to Galton-Watson processes
In practice, if a population is very large (but not infinite), the early generations proceed as if it were in the infinite-population limit.  As long as the total number of infections that have occurred is very small compared to the population size then $S/N$ is very close to $1$.  It is reasonable to assume that each transmission successfully causes an infection.  However, once the number infected is large enough that we expect that some individuals have received multiple transmissions, we must account for the fact that not all transmissions cause new infections.
