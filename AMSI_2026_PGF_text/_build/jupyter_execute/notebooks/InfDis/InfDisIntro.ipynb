# Infectious Diseases 

## Introduction
We will consider directly-transmitted infectious diseases.  Interactions between infectious and susceptible individuals can lead to infection of the susceptible individual.

There are two general classes of infectious disease that are typically studied: 
- **SIS Diseases** for which susceptible individuals can become infected, and later recover and are susceptible to infection again.
- **SIR Diseases** which transmit similarly to SIS diseases, but once an individual recovers he/she gains permanent immunity.

There are a number of assumptions that we will typically make about infectious diseases:

- We usually assume the number of transmissions caused by an infectious individual is a random number, and its distribution is the *offspring distribution*.  We will typically use $\mu(x)$ as the PGF of the offspring distribution.  Note that if some recipients of transmissions are not susceptible, not all transmissions will cause infections.

- We often assume the population is *well-mixed*, that is, the recipient of any transmission is randomly selected from the population.

- We will often do calculations in the *infinite-population limit*, corresponding to the limiting behaviour as the population size $N \to \infty$.




In practice, if a population is very large (but not infinite), the early generations proceed as in the infinite-population limit.  As long as the total number of infections that have occurred is very small compared to the population size, it is reasonable to assume that each transmission causes an infection.  However, once the number infected is large enough that we expect that some individuals have received multiple transmissions, then we must account for the fact that not all transmissions cause new infections.
