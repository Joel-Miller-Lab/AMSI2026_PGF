# Size Distribution at successive times

Please read {numref}`sec:GenFxnLinearComb`, {numref}`sec:GenFxnMult`, and {numref}`sec:GenFxnComp` before reading this section.

The major result we rely on in thiss section is that if $\mu_1(x)$ and $\mu_2(x)$ are both PGFs, then $\psi(x) = \mu_1(\mu_2(x))$ is the PGF for the following process:
First we choose $k$ from the distribution of $\mu_1$.  Then we choose $k$ numbers from the distribution of $\mu_2$ and add them together.

This can be used to prove the following theorem:

```{prf:theorem} Galton-Watson sizes at later generations
:label: theorem-GW-size-by-generation

If $\mu(x)$ is the PGF of the offspring distribution of a Galton-Watson process, then the distribution of sizes at generation $g$ has PGF $\mu^{(g)}(x)$ where $\mu^{(0)}(x) = x$ and $\mu^{(g)}(x) = \mu^{(g-1)}(\mu(x)) = \mu(\mu^{(g-1)}(x))$.
```

We will give two proofs of this theorem.  At first glance the difference between the two proofs appears trivial.  However, later when we will look at continuous-time variations of the Galton-Watson process the two methods will lead to very different equations.

```{prf:proof} "Forward Proof"
:label: proof-GenSizes-forward

We use induction on the generation $g$.  At the base case, $g=0$, we know that $X_0=1$, so $\mu^{(0)}(x)=x$.  

For the inductive step, we know that the distribution of $X_{g-1}$ is $\mu^{(g-1)}(x)$.  By the definition of the Galton-Watson process, $X_g$ is found by choosing $X_{g-1}$ numbers from the offspring distribution and adding them together.  So the PGF of the distribution of $X_g$ is $\mu^{(g-1)}(\mu(x))$.
```
