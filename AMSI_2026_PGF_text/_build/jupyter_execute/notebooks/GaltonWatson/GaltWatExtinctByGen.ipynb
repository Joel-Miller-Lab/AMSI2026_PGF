# Extinction probability over time

In {numref}`sec:GaltWatExtinct`, we showed that the probability $\alpha$ of extinction of a Galton-Watson Process is a solution to the equation $x = \mu(x)$.  However, we saw that sometimes there are two solutions in $[0,1]$, so identifying which is the correct solution may not be obvious.

To find the correct solution, we calculate $\alpha_g$, the probability that it is extinct by generation $g$.  We assume one inividual exists in generation $g=0$, and thus $\alpha_0=0$.  In order for it to go extinct by generation $g=1$, the initial infected individual must infect no-one, so $\alpha_1 = p_0 = \mu(0)$.

For $g>1$, we follow a very similar argument to our initial argument that showed $\alpha = \mu(\alpha)$.  The probability that the process goes extinct by generation $g$ is equal to the probability that all direct offspring of the initial individual spark their own Galton-Watson Process that is extinct within $g-1$ generations.
