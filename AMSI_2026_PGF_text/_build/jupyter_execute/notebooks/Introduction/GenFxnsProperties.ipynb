# Properties of Generating functions

As noted previously, we will be interested in the values of 

$$f(x)=c_0 + c_1x + c_2 x^2 + \cdots$$ 
when it contains information about the sequence of coefficients of $f$.

A few immediate properties can be observed:

- $c_0 = f(0)$
- $c_0 + c_1 + c_2 + \cdots = f(1)$
- $c_1 + 2c_2 + \cdots = f'(1)$

(these only hold if the sums converge - if $f(x)$ has a singularity between $0$ and $1$, the sums do not converge).


## Properties of PGFs

Consider a probability distribution $p_0, p_1, \ldots$.  Let $f(x)$ be the corresponding PGF.  Then

- $f(0)=p_0$
- $f(1) = 1$
- $f'(1) = \sum k p_k = \langle K \rangle$ where $\langle K \rangle$ denotes the average value of $k$.
- $f''(1) = \sum k(k-1)p_k = \langle K^2-K\rangle$.
- $\left. \frac{d}{dx} \left(x f'(x)\right) \right|_{x=1} = \langle K^2 \rangle$.

These can all be proven by writing out the summations and substituting in $x=1$.
