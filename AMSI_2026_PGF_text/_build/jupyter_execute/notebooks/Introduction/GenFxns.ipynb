# Generating Functions and simple biological processes

In this unit, we will study functions that can be expressed as a power series:

$$
\mu(x) = p_0 +p_1 x + p_2 x^2 + \cdots
$$
where $p_k \geq 0$ can be thought of as the probability of $k$.  For our purposes we will usually be interested in the specific case where $p_k$ is the probability one organism creates $k$ offspring.

By performing mathematical operations with the function $\mu(x)$ we will learn how to calculate things such as
- The probability of extinction within $g$ generations,
- The expected number of organisms after $g$ generations,
- The probability of exactly $n$ organisms after $g$ generations,
and many related quantities of interest.

This first chapter will alternate between discussion of the mathematical properties of generating functions (especially probability generating functions) and the Galton-Watson process which is frequently used to study populations in which there is a known distribution of the number of offsprings an individual has.
