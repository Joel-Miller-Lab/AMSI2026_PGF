# Total Size of a (finite) Galton-Watson Process

Consider a Galton-Watson Process whose offspring distribution has PGF $\mu(x)$.  It is useful to know the probability that the process terminates after a cumulative size of exactly $j$.

```{prf:theorem} Distribution of Final Sizes of Galton-Watson Processes
:label:theorem-TotalSizeDist

Given a Galton-Watson Process with offspring distribution PGF of $\mu(x)$, the that the process terminates after a cumulative count of exactly $\sum_{g=0}^\infty X_g = j$ is
given by $\frac{1}{j} p_{j-1}^{(j)}$ where $p_{j-1}^{(j)}$ denotes the coefficient of $x^{j-1}$ of $[\mu(x)]^j$
```
