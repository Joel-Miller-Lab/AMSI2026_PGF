# The Backward Kolmogorov Equation

The backward Kolmogorov equation along similar lines to the forward Kolmogorov equation.  We assume we know $\Phi(x,t)$ and $\mu(x)$ and we want to determine $\Phi(x, t + \Delta t)$.

However, instead of taking time $t$ and stepping forward an amount $\Delta t$, we take time $\Delta t$ and step forward an amount $t$.

Given $\Phi(x,0)= x$, that is, there is one individual at time $t=0$, we can find $\Phi(x,\Delta t)$.  With probability $(1-r\Delta t)$ nothing has happened, but with probability $r\Delta t$, an event has occured and the new PGF is $\mu(x)$.  So

$$
\Phi(x,\Delta t) = (1-r\Delta t)x + r\Delta t \mu(x)
$$

By the rules of composition of PGFs, the number at time $t+\Delta t$ is given by $\Phi(\Phi(x,t),\Delta t)$.  So

$$
\Phi(x,t+\Delta t) = (1-r\Delta t) \Phi(x,t) + r \Delta t \mu(\Phi(x,t))
$$
Calculating the derivative

\begin{align*}
\frac{\partial}{\partial t} \Phi(x,t) &= \lim_{\Delta t \to 0} \frac{\Phi(x,t+\Delta t)-\Phi(x,t)}{\Delta t}\\
&=\lim_{\Delta t \to 0} \frac{(1-r\Delta t) \Phi(x,t) + r \Delta t \mu(\Phi(x,t)) - \Phi(x,t)}{\Delta t}\\
&= -r\Phi(x,t) + r \mu(\Phi(x,t)) \\
&= r (\mu(\Phi(x,t))-\Phi(x,t))
\end{align*} 

This yields the Backward Kolmogorov Equation

$$
\frac{\partial}{\partial t} \Phi(x,t) = r (\mu(\Phi(x,t))-\Phi(x,t))
$$

```{prf:example} Extinction probability by time $t$
:label: example-BackKolmExtProb

The probability of extinction by time $t$ is $q_{0}(t) = \Phi(0,t)$.  We define this to be $\alpha(t)$.  The Backward Kolmogorov Equation gives

\begin{align*}
\alpha'(t) &= \left[\frac{\partial}{\partial t} \Phi(x,t)\right|_{t=0}\\
&= r(\mu(\Phi(0,t)-\Phi(0,t)))\\
&= r(\mu(\alpha(t))-\alpha(t))
\end{align*}
with the initial condition $\alpha(0)=0$.
```

## Self-test
