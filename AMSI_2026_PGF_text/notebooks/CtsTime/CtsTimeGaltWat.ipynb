
# Continuous Time Galton Watson Process

We now assume that individuals reproduce or die as Poisson processes.  Individuals reproduce with one rate (having a random number of offspring) and die with another rate.

Conceptually if all we care about is the counts at time $t$, one individual creating $k-1$ offspring is equivalent to one individual being replaced by $k$ offspring.  This allows us to combine our reproduction and deaths into a single Poisson process with a combined rate $r$.  When the event occurs, the individual is removed and replaced by number of offspring with a given distribution.  We define

$$\hat{\mu}(x) = p_0 + p_1x + p_2x^2\cdots $$
Typically we assume $p_1 \neq 0$.  For most infectious disease models we assume that only $p_0$ and $p_2$ are nonzero.


Note that if we want to count the cumulative number of individuals, we cannot assume that the individual is removed and then replaced.  Different methods will be needed then.
