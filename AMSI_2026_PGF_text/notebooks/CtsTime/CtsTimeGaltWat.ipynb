
# Continuous Time Galton Watson Process

We now assume that we begin with a single individual and individuals reproduce or die as Poisson processes.  Individuals reproduce with one rate (having a possibly random number of offspring) and die with another rate.

Conceptually if all we care about is the counts at time $t$, one individual creating $k-1$ offspring is equivalent to one individual being removed and replaced by $k$ offspring.  This allows us to combine our reproduction and deaths into a single compound Poisson process with a combined rate $r = \sum_k r_k$ where $r_k$ is the probability of being replaced by $k$ offspring.

This leads to the following definition:

```{prf:definition} Continuous-Time Galton-Watson Process
:label: def-CTGW-XFocused

A **continuous-time Galton-Watson process** is defined as follows.

- The process begins with $X(0)=1$.
- $X$ changes by discrete jumps at time $t_1$, $t_2$, $\ldots$, with the time between $t_i$ and $t_{i+1}$ exponentially distributed with total rate $X(t_i) r$ (taking $t_0=0$).
- At the jumps, $X$ changes by an amount $k-1$ with probability $p_k = r_k/r$ (so $\sum_k r_k = r < \infty$).
```
We can create a definition resembling the Galton-Watson Tree definition.  
```{prf:definition} Continuous-Time Galton Watson Process
:label: def-CTGW-treelike
A **continuous-time Galton-Watson process** is defined as follows.

- The process begins with a single individual at $t=0$.
- Each individual lives for an exponentially distributed time with total rate $r = \sum_{k\geq 0} r_k < \infty$, independently of all other events.
- Upon death, she is instantaneously replaced by $k$ individuals with probability $r_k/r$, independently of all other events.
```



Using $p_k=r_k/r$ to be the probability that after an event, the individual has been replaced by $k$ individuals, we define

```{prf:definition} The Offspring PGF $\hat{\mu}(x)$
:label: def-CtsOffspringPGF
We will use the PGF

$$ 
\hat{\mu}(x) = p_0 + p_1x + p_2x^2\cdots 
$$
to be the PGF of the offspring distribution in the continuous time Galton-Watson process.
```

```{prf:definition} $X(t)$
:label: def-CtsTimeX

We use $X(t)$ to denote the (stochastic) population size at time $t$.
```

If all we care about is calculating $X(t)$, we can assume $p_1 \neq 0$.  For most infectious disease models we will assume that only $p_0$ and $p_2$ are nonzero, but for other processes we can have $p_k >0$ for $k >2$.

Note that if we want to count the cumulative number of individuals in our real-world application, we cannot assume that the individual is removed and then replaced when giving birth.  We would need to be careful with the accounting in this case.


## PGF at small times

It will be important to know what the PGF of $X(\Delta t)$ is when $\Delta t \ll 1$.

We will ignore small terms in the derivation below.  They could be kept and later shown to go to zero.

We begin with one individual at $t=0$.  The probability that the individual dies during the first $\Delta t$ time units is (approximately) $r \Delta t$.  If so, the number of individuals replacing her has PGF $\hat{\mu}(x)$.  If not, then the resulting PGF is simply $x$.

For sufficiently small $\Delta t$, we can assume that at most one event occurs in the time interval.  Then taking the appropriate convex combination of $\hat{\mu}(x)$ and $x$, we find

\begin{align*}
\Phi(x,\Delta t) &= (1-r\Delta t) x + (r \Delta t) \hat{\mu}(x)\\
 &= x + (r \Delta t) (\hat{\mu}(x)-x)
\end{align*}

We turn this into a Lemma for future use, holding on to the small error terms that were neglected above:

```{prf:lemma}  PGF of $X(t)$ at time $\Delta t$
:label: lem-PGFatDt

If $X(0)=1$, the PGF of the population size $X(t)$ at time $\Delta t \ll 1$ is

\begin{align*}
\Phi(x,\Delta t) &= (1-r\Delta t) x + (r \Delta t) \hat{\mu}(x) + \mathcal{o}(\Delta t)\\
 &= x + (r \Delta t) (\hat{\mu}(x)-x)+ \mathcal{o}(\Delta t)
\end{align*}
```

This is illustrated in {numref}`fig-PhixDt`

```{figure} Phi_x_Dt.png
---
width: 600px
name: fig-PhixDt
---
Illustration of {prf:ref}`lem-PGFatDt`.
```

## Mapping to the generation-based Galton-Watson process
In defining $\hat{\mu}(x)$, we assume that when an event happens an individual either dies (with probability $p_0$) or gives birth to $k-1$ individuals (with probability $p_k$, $k>1$), but we have treated it as if the individual dies and is replaced by $k$ individuals.  This assumption simplifies the calculations, but if we want to relate our results in the continuous time case to the generation-based case, we need to undo this assumption.

If we assume that the individual does not die in the birth events, then conditional on a birth event, the probability of having $k$ offspring is $q_k = p_{k+1}/(1-p_0)$.  The corresponding PGF is:

\begin{align*}
\sum_k q_k x^k &= \sum_k \frac{p_{k+1}}{1-p_0} x^k\\
&= \frac{p_2}{1-p_0} x + \frac{p_3}{1-p_0} x^2 + \cdots\\
& = \frac{ \hat{\mu}(x) - p_0}{(1-p_0)x}
\end{align*}
The PGF for the number of birth events $m$ comes from the geometric distribution, {prf:ref}`example-GeomPGF` (not counting the final event) is $p_0/(1-(1-p_0)x)$.

Since the offspring distribution for the entire lifetime is a randomly stopped sum of $m$ terms each with PGF $\frac{ \hat{\mu}(x) - p_0}{(1-p_0)x}$, we have

\begin{align*}
\mu(x) &= \frac{p_0}{1-(1-p_0)\frac{ \hat{\mu}(x) - p_0}{(1-p_0)x}}\\
   &= \frac{x p_0}{x - (\hat{\mu}(x)-p_0)}
\end{align*}
