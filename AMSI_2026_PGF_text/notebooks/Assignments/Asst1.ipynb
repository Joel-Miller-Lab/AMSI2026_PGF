   **(b)** Explain how the generating-function paradigm interprets $f(x)$ as a compact way of encoding this sequence. In your explanation, describe how the coefficients $c_n$ can be recovered from the function using derivatives evaluated at $x = 0$.

4. In high school I came across a method that seemed to find all of the numbers that are both triangular numbers $(m)(m+1)/2$ and square numbers $n^2$ (I couldn't prove it at the time).  

   Consider the sequence $a_n$ defined by $a_0=1$, $a_1=2$ and $a_{n+2} = a_n + 2a_{n+1}$ and the sequence $b_n$ defined by $b_0=1$, $b_1 = 3$ and $b_{n+2} = b_n + 2b_{n+1}$.  Then $(a_nb_n)^2$ is both triangular and square.

   I've since learned that this property of the sequence is known, but it's still an interesting result.  We can use OGFs to prove that these are both triangular and square (though it doesn't prove that every triangular and square number is in the sequence).  Some of these steps follow our derivation of the formula for the Fibonacci numbers.

   
   **(a)** For $n=0, 1, 2$ use direct calculate to verify $b_n^2 + (-1)^n = 2a_n^2$.  Thus it follows that $(a_nb_n)^2 = b_n^2(b_n^2\pm 1)/2$ is both square and triangular.  Now we want to show this for the general case, and be able to quickly find the $n$-th such number.

   **(b)** Derive the OGFs $f(x)$ and $g(x)$ for both $a_n$ and $b_n$.  

   **(c)** Let $Q= 1+\sqrt{2}$ and $R = 1-\sqrt{2}$.  Your denominators for the OGFs should equal $(1-Qx)(1-Rx)$.  Write $f(x) = A/(1-Qx) + B/(1-Rx)$.  Find $A$ and $B$.
   
   **(d)** Use this to write $a_n$ in terms of $Q$ and $R$. Similar steps show that $b_n = (Q^{n+1} + R^{n+1})/2$.
   
   **(e)** Write $b_n^2$ in terms of $Q$ and $R$, simplifying any $QR$ term using $QR=-1$.
   
   **(f)** Similarly simplify $2a_n^2$ in terms of $Q$ and $R$, and show that $2a_n^2  = b_n^2 + (-1)^n$.  
   
   **(g)** Thus explain why $b_n^2 a_n^2$ is a triangular number (it is clearly square).
