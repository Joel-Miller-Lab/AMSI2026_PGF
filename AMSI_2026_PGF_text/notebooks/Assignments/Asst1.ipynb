3. Let $X$ be a discrete random variable with values $\{0,1,2,\ldots\}$ and 

   $$
   \mathbb{P}[X = k] = C \left(\frac{1}{2}\right)^k \quad \text{for } k = 0,1,2,\ldots
   $$

   where $C$ is a normalizing constant.

   **(a)** Determine the value of $C$.

   ```{dropdown} Solution (a)
   We need the sum of the probabilities to be $1$, so

   \begin{align*}
   C &= \left( \sum_k \left(\frac{1}{2}\right)^k)^{-1}\\
     &= \left( \frac{1}{1-1/2}\right)^{-1}\\
     &= 2^{-1}\\
     &= 1/2
   \end{align*}
   ```

   **(b)** Find the probability generating function $\mu(x)$ in closed form.

   ```{dropdown} Solution (b)

   \begin{align*}
   \mu(x) &= \sum_k \mathbb{P}[X=k]x^k\\
          &= \sum_k \frac{1}{2} \left(\frac{1}{2}\right)^kx^k\\
          &= \frac{1}{2}\frac{1}{1-x/2}\\
          &= \frac{1}{2-x}
   \end{align*}
   ```


   **(c)** Use $\mu(x)$ and $x \partial_x$ to find $\sum_k k^2/2^k$.

   ```{dropdown} Solution (c)
   Applying $x \partial_x$ to a generating function $\sum_k c_kx^k$ yields $\sum_k k c_k x^k$.  Applying it twice yields $\sum_k k^2 c_k x^k$.  So:

   \begin{align*}
   \sum_k C\frac{k^2}{2^k} x^k&=(x\partial_x)^2 \mu(x)\\
    &= x\partial_x \left(x \partial_x (2-x)^{-1}\right) \\
    &= x \partial_x \left(x (-1)(2-x)^{-2}(-1)\right)
    &= x \partial_x x(2-x)^{-2}\\
    &= x\left( (2-x)^{-2} + x(-2)(2-x)^{-3}(-1)\right)\\
    &= \frac{x}{(2-x)^2} + \frac{2x^2}{(2-x)^3}
    \end{align*}
    So evaluating at $x=1$ gives $C\sum k^2/2^k = 1/1^2 + 2/1^3 = 3$.  Since $C=1/2$, we conclude that

    $$
    \sum_k k^2/2^k = 6
    $$
   ```

5. 
   **(a)** Show how the variance of a distribution can be written using derivatives of a PGF evaluated at $1$.

   ```{dropdown} Solution (a)
   The variance is $\mathbb{E}[X^2] - \mathbb{E}[X]^2$.  For the PGF $\mu(x)$, we have

   \begin{align*}
   \mu'(1) &= \sum_k k p_k 1^{k-1}\\
       &= \mathbb{E}[X]
   \end{align*}
   and
   \begin{align*}
   \mu''(1) &= \sum_k k(k-1) p_k 1^{k-2}\\
       &= \sum_k k^2 p_k - \sum_k k p_k
       &= \mathbb{E}[X^2]-\mathbb{E}[X]
   \end{align*}

   So $\mu''(1) + \mu'(1) - \mu'(1)^2 = \mathbb{E}[X^2] - \mathbb{E}[X]^2$, which is the variance.
   ```

   **(b)** Show that the variance equals the mean, iff $\mu''(1)/\mu'(1) = \mu'(1)$.  

   ```{dropdown} Solution (b)
   If variance equals the mean then

   \begin{align*}
   \mu''(1) + \mu'(1) - \mu'(1)^2 &= \mu'(1)\\
   \Rightarrow \mu''(1) -\mu'(1)^2&=0\\
   \Rightarrow \mu''(1) &= \mu'(1)^2\\
   \Rightarrow \frac{\mu''(1)}{\mu'(1)} &= \mu'(1)
   \end{align*}

   In the other direction, if $\mu''(1)/\mu'(1) = \mu'(1)$, then we simply note that each step in the sequence of equations above is reversible.  So the opposite direction follows as well.
   ```

   **(c)** Use PGFs to show that the variance of the Poisson distribution equals its mean.

   ```{dropdown} Solution (c)
   The Poisson distribution with mean $\lambda$ has PGF
   
   $$
   \mu(x) = e^{\lambda(x-1)}
   $$
   Taking $\mu''(1)/\mu'(1)$ we have $\lambda^2/\lambda = \lambda$

   ```

7. Consider a pair of standard tetrahedral (4-sided) dice numbered $1, 2, 3, 4$.

   **(a)** Find the PGF of the sum found by rolling both.

   ```{dropdown} Solution (a)
   The PGF for one die is $\psi(x)=(x+x^2+x^3+x^4)/4$.  The PGF for the sum of two dice is found by multiplying the PGFs together:

   \begin{align*}
   \psi(x)^2 &= \frac{(x+x^2+x^3+x^4)^2}{16}\\
    &= \frac{x^2 + 2x^3 + 3x^4 + 4x^5 + 3x^6+2x^7+x^8}{16}
   \end{align*}

   ```

   **(b)** Find another pair of 4-sided dice that have positive integer values and give the same sums with the same probabilities.

   ```{dropdown} Solution (b)
   We can factor 
   \begin{align*}
   \psi(x) &= \frac{x+x^2+x^3+x^4}{4}\\
           &= \frac{x(1+x+x^2+x^3)}{4}\\
           &= \frac{x(1+x)(1+x^2)}{4}
   \end{align*}
   So $\psi(x)^2 = x^2(1+x)^2(1+x^2)^2/16$.

   Looking for $\mu_1(x) = \frac{x^{a_1}+ x^{a_2}+x^{a_3}+x^{a_4}}{4}$ and $\mu_2(x)=\frac{x^{b_1}+ x^{b_2}+x^{b_3}+x^{b_4}}{4}$ so that their product is $\psi(x)^2$ and all $a_i$ and $b_i$ are positive integers.  
   
   We must have

   \begin{align*}
   \mu_1(x) &= \frac{x^{q_1}(1+x)^{q_2}(1+x^2)^{q_3}}{4}\\
   \mu_2(x) &= \frac{x^{1-q_1}(1+x)^{1-q_2}(1+x^2)^{1-q_3}}{4}
   \end{align*}
   where each $q_i$ is $0$, $1$, or $2$.

   We have some constraints that we can use to idenitfy $q_i$.

   - As the exponents are positive integers, we must have $\mu_1(0)=\mu_2(0)=0$.  So $q_1=1$.
   - At $x=1$, both must be $1$.  So $q_2+q_3=2$.  
   
   If $q_2=q_3=1$, then we recover the original pair.  Otherwise we can assume $q_2=2$ and $q_3=0$ (alternately $q_2=0$ and $q_3=2$ simply interchanges $\mu_1$ and $\mu_2$).  So

   \begin{align*}
   \mu_1(x) &= \frac{x(1+x)^2}{4}\\
            &= \frac{x+2x^2+x^3}{4}\\
   \mu_2(x) &= \frac{x(1+x^2)^2}{4}\\
            &= \frac{x+2x^3+x^5}{4}
   \end{align*}
   So the dice are numbered $1,2,2,3$ and $1,3,3,5$.
   ```

   **(c)** Show that the only possible pairs are the standard dice and the pair you have just found.

   ```{dropdown} Solution (c)
   According to our constraints found in the previous solution, there is no other way to create the pair, and thus this is unique.  (as a more subtle point, this relies on the fact that factorization of polynomials with real coefficients is unique).
   ```

    ```{dropdown} Solution (a)

    $$
    ```
