    **(e)** Show that $\Xi(1,t) = 1-\alpha(\infty)$.

    **(f)** Show that $\lim_{t\to\infty} \lim_{x \to 1} \Omega(x,t) \neq \lim_{x\to 1} \lim_{t\to\infty} \Omega(x,t)$ (if $\alpha(\infty)<1$).
    
    The family of functions $\Omega(x,t)$ for $t=1,2,\ldots$ is a family of functions that are all continuously differentiable, their limit exists for all $x \in [0,1]$, but the limit is not even continuous.  The jump is due to probability mass that moves to larger and larger $k$ over time, making $x^k$ go to zero for any $|x|<1$.  Mathematically, $\lim_{t\to\infty} \Omega(x,t) = \sum_k \omega_k(\infty)x^k$ for $|x|<1$ and $1$ for $x=1$ is not a PGF.  We can patch it up somewhat by introducing $\omega_\infty x^\infty$ where $\omega_\infty = 1-\alpha(\infty)$ is the probability the outbreak grows to unbounded size and $x^\infty$ is assumed to be $0$ for $0\leq x < 1$ and $1$ for $x=1$.
