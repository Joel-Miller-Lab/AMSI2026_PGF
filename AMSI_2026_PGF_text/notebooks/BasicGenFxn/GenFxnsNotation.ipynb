# Some notation

There are a few pieces of notation that we will try to use consistently.  There is potential for confusion as in other branches of mathematics some of this notation would have a different interpretation.

- We will use the Latin alphabet for Ordinary Generating Functions and the Greek alphabet for Probability Generating Functions.
- We will frequently need to reference the coefficient of a particular power of $x$ in the series expansion of $f(x)$.  For this, we will use 

  $$
  [x^n]f(x)
  $$

For example if $f(x)=(1+x)^2=1+2x+x^2$, then $[x]f(x) = 2$.


- Function composition will occur frequently.  We will adopt a notation that is often used for differentiation in other areas of mathematics.  Given the function $f(x)$, the function $f^{(2)}(x)$ is $f(f(x))$.  More generally 

$$f^{(n+1)}(x) = f(f^{(n)}(x))$$
- Multiplication of functions will also occur frequently.  We use $f(x)^2$ to represent $f(x)f(x)$.  More generally 

$$f(x)^{n+1} = f(x)f(x)^n$$
- Derivatives occur often.  Usually we will use the standard notation $\frac{d^n}{dx^n}f(x)$ to denote the $n$th derivative, and if $n$ is small we will often write $f''(x)$, but never $f^{(n)}(x)$ as this is reserved for function composition.  
- We are often interested in $xf'(x)$.  We will use the notation $x \partial_x f(x)$ to denote $x f'(x)$ and 

$$(x \partial_x)^{n+1} f(x)= x \partial_x (x\partial_x)^n f(x)$$


|Notation|Interpretation|
|--|--| 
|$f(x), g(x), \ldots$ | Ordinary Generating Functions|
|$\mu(x), \psi(x), \ldots$ | Probability Generating Functions|
|$[x^n]f(x)$, $[x^n]\mu(x)$| The coefficient of $x^n$ in the series expansion of $f(x)$ or $\mu(x)$|
|$f^{(n)}(x)$, $\mu^{(n)}(x)$| function composition: $f(f(f(\cdots f(x) \cdots)))$ or $\mu(\mu(\mu(\cdots \mu(x)\cdots)))$|
|$f(x)^n$, $\mu(x)^n$| function multiplication: $f(x)f(x)\cdots f(x)$ or $\mu(x)\mu(x)\cdots \mu(x)$|
|$(x \partial_x)^n f(x)$, $(x \partial_x)^n \mu(x)$|  $x\frac{d}{dx}\left(x \frac{d}{dx}\left(\cdots x\frac{d}{dx} f(x)\cdots\right)\right)$ or $x\frac{d}{dx}\left(x \frac{d}{dx}\left(\cdots x\frac{d}{dx} \mu(x)\cdots\right)\right)$
