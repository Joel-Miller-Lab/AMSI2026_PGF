# Multivariate Probability Generating Functions

We now consider PGFs that depend on multiple variables

Let us assume that we have some distribution joint distribution of two non-negative integer random variables $p_{i,j}$.  Then we define a two-variable PGF

$$
\mu(x,y) = \sum_{i=0}^\infty \sum_{j=0}^\infty p_{i,j} x^i y^j
$$
