# Generation-Based Infectious Disease Models

When we switch to a generation-based framework, we allow a wider range of offspring distributions.  Now each infected individual transmits to $k$ others, where $k$ is a random variable and the recipients are chosen randomly from the population.  If the recipient is susceptible, he becomes infected.  Otherwise, it has no effect.

As before, while $S/N$ is approximately $1$, we can treat this as if it is a generation-based Galton-Watson process.  Thus all of our results for Galton-Watson processes apply to $I(g)$.


## Size distribution of small outbreaks

The derivation we did showing that the probability a tree has $j$ nodes is

$$ 
\mathbb{P}[\text{tree has $j$ nodes}] = \frac{1}{j} \left[x^{j-1}\right] \left(\mu(x)^j\right)
$$
applies immediately here.  The only difference is that for this model, each node represents a distinct infected individual.  Thus $j$ is the size of the small outbreak.


```{prf:example} Continuous-time case from a generation-based perspective

Consider the continuous-time SIS or SIR model with rates $\beta$ and $\gamma$. We will convert this to a generation-based framework and use this to calculate the probability distribution of outbreak sizes. 

The probability the next event from an individual is a recovery is $p = \gamma/(\beta+\gamma)$ and the probability of transmission is $1-p$.  The probaiblity of exactly $k$ transmissions before the recovery is 

$$
p_k = p(1-p)^k
$$
By looking at the PGF for a geometric distribution, we could find $\mu(x)$, but for practice we calculate it directly

\begin{align*}
\mu(x) &= \sum_k p_k x^k\\
  &= p \sum_k (x(1-p))^k\\
  &= \frac{p}{1-x(1-p)}
\end{align*}
Thus the probability of exactly $j$ infections in the outbreak is

\begin{align*}
\mathbb{P}[j \text{ infections}] &= \frac{1}{j} \left[x^{j-1}\right] \left(\frac{p}{1-(1-p)x}\right)^j\\
&= \frac{1}{j} \left[x^{j-1}\right]\left(\sum_{N=0}^\infty \binom{N+j-1}{N} p^j (1-p)^N x^N\right)\\
&= \frac{1}{j} \binom{2j-2}{j-1} p^j (1-p)^{j-1}
\end{align*}
(we used the expansion for the negative binomial distribution here, from {prf:ref}`example-NegBinPGF`). Substituting $p=\gamma/(\beta+\gamma)$ we find

\begin{align*}
\mathbb{P}[j \text{ infections}] &= \frac{1}{j} \binom{2j-2}{j-1} \frac{\beta^j\gamma^{j-1}}{(\beta+\gamma)^{2j-1}}\\
&= \frac{1}{j} \binom{2j-2}{j-1} \frac{\mathcal{R}_0^j}{(1+\mathcal{R}_0)^{2j-1}}
\end{align*}
which is what we derived for the continuous-time case.
```
